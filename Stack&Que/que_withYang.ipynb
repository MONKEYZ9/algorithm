```python
class Queue:
    def __init__(self, n):
        self.array = [None for _ in range(n)]
        self.f_idx = 0
        self.b_idx = 0

    def push(self, num):
        print(self.array[self.b_idx])
        self.array[self.b_idx] = num
        self.b_idx += 1

    def pop(self):
        # if self.f_idx == self.b_idx:
            #  이 둘이 같다면 인자가 없다는 것
        # if self.b_idx - self.f_idx == 0:
        # 이것과 같은 소리다
        if self.is_empty():
            return -1

        pop_val = self.array[self.f_idx]
        self.f_idx += 1
        return pop_val

    def size(self):
        # 백 인덱스 에서 앞 인덱스를 뺀거를 사이즈로 보내면 된다는거지
        return self.b_idx - self.f_idx

    # 명령어로 메소드 명을 하기보다는 is_empty로 하자
    # snake case로 is_empty로 쓰는 거
    # camel case는 isEmpty로 쓰는 거
    
    def empty(self):
        return int(self.is_empty())

    def is_empty(self):
        return self.size() == 0

    def front(self):
        if self.is_empty():
            return -1
        return self.array[self.f_idx]

    def back(self):
        if self.is_empty():
            return
        return self.array[self.b_idx - 1]

def run_cmd_with_que(command, queue_obj):
    cmd_type = command[0]

    if cmd_type=='push':
        # 뭘 푸쉬할 것인가
        _, num = command
        queue_obj.push(int(num))

    elif cmd_type=='pop':
        print(queue_obj.pop())

    elif cmd_type=='size':
        print(queue_obj.size())

    elif cmd_type=='empty':
        print(queue_obj.empty())

    elif cmd_type=='front':
        print(queue_obj.front())

    elif cmd_type=='back':
        print(queue_obj.back())

n = int(input())
queue_obj = Queue(n)

for _ in range(n):
    run_cmd_with_que( input().split(),  queue_obj)

    print(queue_obj.b_idx)
    print(queue_obj.f_idx)
    print(queue_obj.array)
```
